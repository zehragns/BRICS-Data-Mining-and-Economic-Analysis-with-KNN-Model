```python
#Correlation matrix by mean
mean_data_2021 = df[df['Year'] == 2021].groupby("BRICS Countries").agg(
    {col: 'mean' for col in df.select_dtypes(include=np.number).columns}
)

# Define the function
def generate_correlation_heatmap(selected_columns):
    filtered_data = mean_data_2021[selected_columns]
    correlation_matrix = filtered_data.corr()

    # Plot the heatmap
    plt.figure(figsize=(10, 8))
    sns.heatmap(correlation_matrix, annot=True, cmap="inferno", fmt=".2f", linewidths=0.5)
    plt.title("Correlation Heatmap for 2021 Averages", fontsize=16)
    plt.tight_layout()
    plt.show()

# Columns to include in the correlation matrix
selected_columns = ["Population_sc", "Gini_Coefficient_sc", "Total GDP ($)_sc",
                    "Per_Capita_Income_sc", "Poverty_Rate_sc", "HDI_sc"]
generate_correlation_heatmap(selected_columns)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score


# Regression Coefficients Analysis
def calculate_regression_coefficients(df):

    # Define dependent and independent variables
    X = df[["Population_sc", "Gini_Coefficient_sc", "Total GDP ($)_sc", "Poverty_Rate_sc", "HDI_sc"]]
    y = df["Per_Capita_Income_sc"]

    # Fit the linear regression model
    model = LinearRegression()
    model.fit(X, y)

    # Get the coefficients
    coefficients = model.coef_
    intercept = model.intercept_

    # Display results
    print("Regression Coefficients:")
    for feature, coef in zip(X.columns, coefficients):
        print(f"{feature}: {coef:.4f}")
    print(f"Intercept: {intercept:.4f}")

    # Calculate R-squared value
    y_pred = model.predict(X)
    r2 = r2_score(y, y_pred)
    print(f"R-squared: {r2:.4f}")

    # Plot actual vs predicted values
    plt.figure(figsize=(10, 6))
    plt.scatter(y, y_pred, alpha=0.7, color="blue")
    plt.plot([y.min(), y.max()], [y.min(), y.max()], 'r--', linewidth=2)
    plt.title("Actual vs Predicted Per Capita Income", fontsize=16)
    plt.xlabel("Actual Values", fontsize=12)
    plt.ylabel("Predicted Values", fontsize=12)
    plt.tight_layout()
    plt.show()

calculate_regression_coefficients(df)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Select relevant normalized features and target
features = df[['Total GDP ($)_sc', 'Population_sc', 'Gini_Coefficient_sc',
               'Poverty_Rate_sc', 'HDI_sc']]
target = df['Per_Capita_Income_sc']

# Create a binary target variable: Low (0) or High (1) income
threshold = target.median()
binary_target = (target > threshold).astype(int)

scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features_scaled, binary_target, test_size=0.3, random_state=42)

# Experiment with different K values to find the best one
best_k = None
best_accuracy = 0

for k in range(3, 11):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)

    if accuracy > best_accuracy:
        best_k = k
        best_accuracy = accuracy

# Train the final model with the best K value
knn_final = KNeighborsClassifier(n_neighbors=best_k)
knn_final.fit(X_train, y_train)
y_pred_final = knn_final.predict(X_test)

# Evaluate the final model
accuracy_final = accuracy_score(y_test, y_pred_final)
error_rate_final = 1 - accuracy_final
conf_matrix = confusion_matrix(y_test, y_pred_final)
class_report = classification_report(y_test, y_pred_final)

# Print the results
print(f"Best K Value: {best_k}")
print(f"Final Model Accuracy: {accuracy_final:.2f}")
print(f"Final Model Error Rate: {error_rate_final:.2f}")
print("Confusion Matrix:")
print(conf_matrix)
print("Classification Report:")
print(class_report)
```
